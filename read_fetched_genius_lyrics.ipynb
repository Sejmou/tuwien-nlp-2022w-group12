# Reading fetched Genius API data

A quick demo of how we can load and analyze the data we fetched from Genius (not all data has been downloaded yet, will update later).

```python
import numpy as np
import pandas as pd
import os
```

```python
data_dir = "data"
lyric_chunk_folder_path = os.path.join(data_dir, "lyric_chunks")
chunk_paths = [os.path.join(lyric_chunk_folder_path, name) for name in os.listdir(lyric_chunk_folder_path)]
chunk_paths.sort()
chunk_paths
```

```python
chunk_paths
```

```python
chunks = [pd.read_json(path, orient="index") for path in chunk_paths]
chunks
```

```python
data = pd.concat(chunks).reset_index()
data
```

```python
data.lyrics.isna().sum()
```

```python
data.lyrics.isna().sum() / len(data)
```

```python
data.language.value_counts()
```
